```ruby
require 'httparty'
require 'nokogiri'
```

```ruby
url = 'https://www.dieese.org.br/notatecnica/2017/notaTecsalariominimo2017.pdf'
page = HTTParty.get(url)
parsed_page = Nokogiri::HTML(page)
tabela = parsed_page.css('table')[1]
linhas = tabela.css('tr')
```

```ruby
linhas.each do |linha|
  next if linha.css('td').empty?

  colunas = linha.css('td')
  
  coluna_programa = linha.css('td', { class: 'coluna_valor'})[0]
  coluna_valor = linha.css('td', { class: 'coluna_valor'})[1]
  
  programa = coluna_programa.text.strip
  valor = coluna_valor.text.strip
  
  puts "#{programa}: #{valor}"
end
```
